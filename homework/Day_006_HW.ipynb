# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
%config IPComplete.greddy=True

# 設定 data_path
dir_data = './data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
# 先篩選數值型的欄位
"""
YOUR CODE HERE, fill correct data types (for example str, float, int, ...)
"""
dtype_select = [np.dtype('float64'), np.dtype('int64')]
numeric_columns = app_train.columns[app_train.dtypes.isin(dtype_select)]

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
not_two_value_columns = app_train[numeric_columns].apply(lambda x: len(x.unique())>2)

numeric_columns = app_train[numeric_columns].columns[not_two_value_columns == True]
print("Numbers of remain columns {:d}".format(len(numeric_columns)))



# 檢視這些欄位的數值範圍
for col in  numeric_columns:
    """
    Your CODE HERE, make the box plot
    """    
    app_train[col].dropna()
    app_train[col].plot.hist(bins=20, title=col)
    
    plt.show()
```

```python
# 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

# AMT_INCOME_TOTAL
# REGION_POPULATION_RELATIVE
# OBS_60_CNT_SOCIAL_CIRCLE

app_train["AMT_INCOME_TOTAL"].plot.hist(title="AMT_INCOME_TOTAL")
plt.show()
app_train["REGION_POPULATION_RELATIVE"].plot.hist(title="REGION_POPULATION_RELATIVE")
plt.show()
app_train["OBS_60_CNT_SOCIAL_CIRCLE"].plot.hist(title="OBS_60_CNT_SOCIAL_CIRCLE")                                           
plt.show()                                        
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""

def calculate_cdf(data):
    data1 = np.sort(data)
    cdf_value1=[]
    for i in data1:
        y1 = data1[data1 <= i].size /data1.size
        cdf_value1.append(y1)
    
    # Theoretical cdf  
    data2 = np.linspace(start=min(data1), stop=max(data1),num=data1.size) 
    cdf_value2=[]
    for i in data2:
        y2 = data2[data2 <= i].size /data2.size
        cdf_value2.append(y2)

    cdf = pd.DataFrame({"CDF":cdf_value1, "Theoretical_CDF":cdf_value2})
    
    return cdf


data = app_train['AMT_INCOME_TOTAL'][:20000]
cdf = calculate_cdf(data)             

cdf.head(20)

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

data.hist()
plt.show()
```
