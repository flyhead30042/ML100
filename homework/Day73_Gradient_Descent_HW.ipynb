```python
cur_x = 3
max_iters = 300 
df = lambda x: 2*(x+5) 

lr=0.1
iters = 0
step_size =999
x_hist_1 = list()

while iters <= max_iters:
    step_size = abs(lr * df(cur_x))
    next_x = cur_x - step_size
    x_hist_1.append(next_x)
    print("[{}]{} <- {}-{}".format(iters, next_x, cur_x, step_size))
    cur_x = next_x
    iters = iters+1    
```

```python
cur_x = 3 
max_iters = 300 
df = lambda x: 2*(x+5) 

lr=0.001
iters = 0
step_size =999
x_hist_2 = list()

while iters <= max_iters:
    step_size = lr * df(cur_x)
    next_x = cur_x - step_size
    x_hist_2.append(next_x)
    print("[{}]{} <- {}-{}".format(iters, next_x, cur_x, step_size))
    cur_x = next_x
    iters = iters+1  
```

```python
import pandas as pd
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,1, figsize=(13,5))
df = pd.DataFrame({"x_0.1":x_hist_1, "x_0.01":x_hist_2})
g = df.plot(kind="line", ax=ax)
```
