# 作業
* 將原始資料集換為S型圖, 觀察不同 perplexity 下的流形還原結果

```python
# 繪製不同 perplexity 下的 t-SNE 流形還原圖 (請注意! 語法也要對應調整)
```

```python
# 設定S曲線資料點
X, color = datasets.samples_generator.make_s_curve(n_samples, random_state=0)

fig, ax = plt.subplots(2,5, figsize=(15,8))

ax[1,4].set_title("Original")
sns.scatterplot(x=X[:, 0], y=X[:, 2], hue=color, ax=ax[1,4], legend=False)

for i, perplexity in enumerate(perplexities):
    tsne = TSNE(n_components=2, random_state=0, init='pca', learning_rate=200., early_exaggeration=12., perplexity=perplexity)
#     tsne = manifold.TSNE(n_components=n_components, init='random', random_state=0, perplexity=perplexity)
    newX = tsne.fit_transform(X)  
    r= int(i/5)
    c = int(i%5)

    ax[r,c].set_title("perplexity={}".format(perplexity))
    sns.scatterplot(x=newX[:, 0], y=newX[:, 1], hue=color, ax=ax[r,c], legend=False)
```

```python
X, color = datasets.make_circles(n_samples=n_samples, factor=.5, noise=.05)

fig, ax = plt.subplots(2,5, figsize=(15,8))

ax[1,4].set_title("Original")
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=color, ax=ax[1,4], legend=False)

for i, perplexity in enumerate(perplexities):
    tsne = TSNE(n_components=2, random_state=0, init='pca', learning_rate=200., early_exaggeration=12., perplexity=perplexity)
    newX = tsne.fit_transform(X)  
    r= int(i/5)
    c = int(i%5)

    ax[r,c].set_title("perplexity={}".format(perplexity))
    sns.scatterplot(x=newX[:, 0], y=newX[:, 1], hue=color, ax=ax[r,c], legend=False)
```
