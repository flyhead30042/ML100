# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pyplot as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
def relu(x):
    return np.maximum(x, 0)
    
def drelu(x):
    y = x > 0
    return y.astype(np.int)
```

```python
import pandas as pd
import seaborn as sns

x = np.linspace(-5,5,100, dtype="float32")
y = relu(x)

df = pd.DataFrame({"x":x, "y":y})
sns.set_style("whitegrid")
g = sns.lineplot(data=df, x="x", y="y")
g.set_title("ReLU")
```

```python
import pandas as pd
import seaborn as sns

x = np.linspace(-5,5,100, dtype="float32")
y = drelu(x)

df = pd.DataFrame({"x":x, "y":y})
sns.set_style("whitegrid")
g = sns.lineplot(data=df, x="x", y="y")
g.set_title("Step_Function")
```
